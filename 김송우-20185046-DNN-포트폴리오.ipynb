````python
    # 0. 데이터 수집
    
    # x 변수, y 변수 데이터
    x = [-3,  31,  -11,  4,  0,  22, -2, -5, -25, -14]
    y = [ -2,   32,   -10,   5,  1,   23,  -1,  -4, -24,  -13]
    
    # 그래프 그리기
    import matplotlib.pyplot as plt
    plt.plot(x, y)
    plt.show()
    
    # 데이터프레임 사용
    import pandas as pd
    df = pd.DataFrame({'X':x, 'Y':y})
    df.shape
    
    df 
````

````python
    # 1. 데이터 전처리
    train_features = ['X']
    target_cols = ['Y']
    X_train = df.loc[:, train_features]  # 데이터프레임에서 X값만 모아 저장
    y_train = df.loc[:, target_cols]     # Y값만 모아서 저장
    print(X_train.shape, y_train.shape)  # 10 rank vector 생성
````

````python
    # 2. 학습(최적의 a값과 b값 구하기)
    from sklearn.linear_model import LinearRegression
    lr = LinearRegression()
    lr.fit(X_train, y_train)   # x_train dataframe과 y_train dataframe으로 학습 
    
    print ("기울기: ", lr.coef_[0][0])   # a(x 계수 )
    print ("y절편: ", lr.intercept_[0])  # b(편향치)
````

````python
    # 3. 모델 예측
    import numpy as np
    X_new = np.array(20).reshape(1, 1)  # 2차원의 1행 1열로 모양 변환
    X_new
    
    lr.predict(X_new)  # 예측 메소드
    
    X_test = np.arange(11, 16, 1).reshape(-1, 1)
    X_test
    
    y_pred = lr.predict(X_test)
    y_pred
````

<ul>
    <li>KNN과 비슷하지만, 모델 학습 부분에서 커널을 지정한다는 점이 다르다.</li>

````python
# 라이브러리 환경
import pandas as pd
import numpy as np

# SVM을 적용한 붓꽃 분류
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn import datasets

# 1. 데이터 준비와 전처리
iris = datasets.load_iris()

# 데이터프레임 생성과 열 지정
df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
df['Target'] = iris['target']

# 중복 데이터 제거
df = df.drop_duplicates()

# 2. 학습 데이터, 테스트 데이터 준비
X_data = df.loc[:, 'sepal_length':'petal_width']
y_data = df.loc[:, 'Target']

X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, 
                    test_size=0.2, shuffle=True, random_state=20)

# 모델 학습
svc = SVC(kernel='rbf')    # 기본적으로 사용할 커널을 지정
svc.fit(X_train, y_train)

# 4. 예측
y_svc_pred = svc.predict(X_test)
print("예측값: ", y_svc_pred[:10])
print("정답 : ", y_test[:10].values)

# 5. 성능 평가
svc_acc = accuracy_score(y_test, y_svc_pred)
print("Accuracy: %.4f" % svc_acc)
````

        <li>손실함수를 최소화하는 가중치W와 편향치b를 구한다.</li>
        <li>가설함수(y와 x간의 관계 유추한 식) : h(x) = Wx+b</li>
        <li>손실함수(실제값과 예측값의 오차) : Cost(W,b) = J = 1/n*sigma(y-h(x))^2  ; MSE</li>
        <li>최적화 과정(손실함수 최소화 방법) : Gradient Decent</li>
            <ul>
                <li>미분값에 따른 기울기의 부호에 따라 극솟값으로 이동</li>

````python
# 텐서플로 2 버전 선택
try:
    # %tensorflow_version only exists in Colab.
    %tensorflow_version 2.x
except Exception:
    pass

# 1 와인 데이터셋 불러오기
import pandas as pd
red = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv', sep=';')
white = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv', sep=';')
print(red.head())
print(white.head())

# 2 와인 데이터셋 합치기
red['type'] = 0
white['type'] = 1
print(red.head(2))
print(white.head(2))

wine = pd.concat([red, white])
print(wine.describe())

# 3 레드 와인과 화이트 와인 type 히스토그램
import matplotlib.pyplot as plt
plt.hist(wine['type'])
# plt.xticks([0, 1])
plt.xticks([0.05, 0.95], ['red', 'white'])
plt.show()

print(wine['type'].value_counts())


import seaborn as sns
# plt.hist(wine['type'], bins=2, rwidth=0.8)
# 히스토그램과 비슷한 Count 그래프. 이산값을 나타낸다. 
ax = plt.subplots()
ax = sns.countplot(x='type', data=wine)
ax = sns.countplot(x='type', data=wine)
ax.set_title('Count of wine types')
ax.set_xlabel('red or white')
ax.set_ylabel('Frequency')


# 4 데이터 요약 정보 확인
print(wine.info())

# 5 데이터 정규화
wine_norm = (wine - wine.min()) / (wine.max() - wine.min())
print(wine_norm.head())
print(wine_norm.describe())

# 6 데이터 섞은 후 numpy array로 변환
import numpy as np
wine_shuffle = wine_norm.sample(frac=1)
print(wine_shuffle.head())
wine_np = wine_shuffle.to_numpy()
print(wine_np[:5])


# 7 train 데이터와 test 데이터로 분리
train_idx = int(len(wine_np) * 0.8)
train_X, train_Y = wine_np[:train_idx, :-1], wine_np[:train_idx, -1]
test_X, test_Y = wine_np[train_idx:, :-1], wine_np[train_idx:, -1]
print(train_X[0])
print(train_Y[0])
print(test_X[0])
print(test_Y[0])
train_Y = tf.keras.utils.to_categorical(train_Y, num_classes=2)
test_Y = tf.keras.utils.to_categorical(test_Y, num_classes=2)
print(train_Y[0])
print(test_Y[0])

# 8 와인 데이터셋 분류 모델 생성
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=48, activation='relu', input_shape=(12,)),
    tf.keras.layers.Dense(units=24, activation='relu'),
    tf.keras.layers.Dense(units=12, activation='relu'),
    tf.keras.layers.Dense(units=2, activation='softmax')
])

model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.07), loss='categorical_crossentropy', metrics=['accuracy'])

model.summary()

# 9 와인 데이터셋 분류 모델 학습
history = model.fit(train_X, train_Y, epochs=25, batch_size=32, validation_split=0.25)

# 10 분류 모델 학습 결과 시각화
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], 'b-', label='loss')
plt.plot(history.history['val_loss'], 'r--', label='val_loss')
plt.xlabel('Epoch')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], 'g-', label='accuracy')
plt.plot(history.history['val_accuracy'], 'k--', label='val_accuracy')
plt.xlabel('Epoch')
plt.ylim(0.7, 1)
plt.legend()

plt.show()

# 5.11 분류 모델 평가
model.evaluate(test_X, test_Y)
````

<li>새로운 등급 new_quality 생성한다.</li>

````python
import matplotlib.pyplot as plt;
import seaborn as sns;

# 11 품질 히스토그램 시각화
sns.countplot(x='quality', data=wine, palette="Set3");
plt.title('Count of wine quality') ;

# 12 품질을 3개의 범주(좋음, 보통, 나쁨)로 재분류
wine.loc[wine['quality'] <= 5, 'new_quality'] = 0
wine.loc[wine['quality'] == 6, 'new_quality'] = 1
wine.loc[wine['quality'] >= 7, 'new_quality'] = 2

print(wine['new_quality'].describe())
print(wine['new_quality'].value_counts())
wine.shape

# 시각화
sns.countplot('new_quality', data=wine, palette="Set3");
plt.title('Count of wine new quality') ;

# 13 데이터 정규화 및 train, test 데이터 분리
del wine['quality']
wine_norm = (wine - wine.min()) / (wine.max() - wine.min())
wine_shuffle = wine_norm.sample(frac=1)
wine_np = wine_shuffle.to_numpy()

train_idx = int(len(wine_np) * 0.8)
train_X, train_Y = wine_np[:train_idx, :-1], wine_np[:train_idx, -1]
test_X, test_Y = wine_np[train_idx:, :-1], wine_np[train_idx:, -1]
train_Y = tf.keras.utils.to_categorical(train_Y, num_classes=3)
test_Y = tf.keras.utils.to_categorical(test_Y, num_classes=3)


````

````python
# 14 와인 데이터셋 다항 분류 모델 생성 및 학습
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=48, activation='relu', input_shape=(12,)),
    tf.keras.layers.Dense(units=24, activation='relu'),
    tf.keras.layers.Dense(units=12, activation='relu'),
    tf.keras.layers.Dense(units=3, activation='softmax')
])

model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.07), loss='categorical_crossentropy', metrics=['accuracy'])

history = model.fit(train_X, train_Y, epochs=25, batch_size=32, validation_split=0.25)

# 15 다항 분류 모델 학습 결과 시각화
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], 'b-', label='loss')
plt.plot(history.history['val_loss'], 'r--', label='val_loss')
plt.xlabel('Epoch')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], 'g-', label='accuracy')
plt.plot(history.history['val_accuracy'], 'k--', label='val_accuracy')
plt.xlabel('Epoch')
plt.ylim(0.7, 1)
plt.legend()

plt.show()
````

    <ul>

        <li>특징추출기</li>
            <ul>
                <li>자동으로 특징을 추출하는 필터를 생성하는 것이 목적, 밑의 2개의 layer가 겹겹이 쌓여있음</li>
                <li>Convolition Layer : 입력 데이터에 필터를 적용 후 활성화 함수를 반영함</li>
                    <ul>
                        <li>필터 집합에 입력 이미지를 통과, output값은 특징값으로 ANN에서 사용</li>
                        <li>필터의 값은 학습을 통해 결정(가중치)</li>
                        <li>매개변수 있음(필터 = 커널 = 윈도우)</li>
                        <img src="https://anhreynolds.com/img/cnn.png" width="50%" height="50%">
                        <img src="https://csdl-images.computer.org/trans/si/2020/04/figures/ko1ab-2961602.gif" width="50%" height="50%">
                        <li>컬러가 있을 경우 3개의 채널을 가짐 - 이미지수,가로,세로,채널/가로,세로,채널,필터수</li>
                        <li>특징맵의 크기가 줄어드는 경우를 방지하기 위해 주변값들을 0으로 채운다.</li>
                        <img src="http://d2l.ai/_images/conv-pad.svg" width="50%" height="50%">
                    </ul>
                <li>Pooling Layer : Subsampling, downsampling 이라고도 부름</li>
                    <ul>
                        <li>네트워크에서 학습해야 하는 매개 변수 수를 줄여 출력을 간소화 (연산량 감소,  이미지의 특징만에 집중)</li>
                        <li>매개변수가 없다</li>
                        <li>과적합 문제 해결</li>
                        <li>데이터의 공간적 특성을 유지할 수 있도록 함 + 이미지 크기 줄임</li>
                            <ul>
                                <li>max pooling(특정 영역 안의 최댓값, 주로 사용하는 풀링)</li>
                                <li>average pooling, min pooling</li>
                            </ul>
                    </ul>
            </ul>
        <li>분류기 : 이미지 분류를 위한 ANN같은 레이어 추가</li>
    </ul>
<li>데이터의 공간적 특성 유지(데이터 평탄화 x ; nx1행렬로 변경하지 않고 그냥 사용)</li>
